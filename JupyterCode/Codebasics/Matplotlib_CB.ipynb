```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=4,linestyle='-.')
plt.show()
```

# Matplotlib Tutorial 3 - Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='upper right',fontsize='large') # default: loc='best'
plt.grid()
plt.show()
```

# Matplotlib Tutorial 4 - Bar Chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue (bln.)')
plt.title('US Tech Stocks')
plt.bar(xpos-.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+.2,profit,width=0.4,label='Profit')
plt.legend()
plt.show()
```

```python
plt.yticks(xpos,company)
plt.xlabel('Revenue (bln.)')
plt.title('US Tech Stocks')
plt.barh(xpos-.2,revenue,label='Revenue')
plt.barh(xpos+.2,profit,label='Profit')
plt.legend()
plt.show()
```

# Matplotlib Tutorial 5 - Histograms

```python

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
# plt.hist(blood_sugar) #default bin 10
# plt.hist(blood_sugar,bins=3,rwidth=0.95)
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='g',histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100] 

plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['Men','Women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['Men','Women'],orientation='horizontal')
plt.legend()
```
