```python
from datetime import datetime 

fig = plt.figure(figsize=(12,8)) 
ax = fig.add_subplot( 1, 1, 1) 

data = pd.read_csv('data_sets/spx.csv', index_col = 0, parse_dates = True)

spx = data['SPX'] 
spx.plot( ax = ax, style ='k-') 
crisis_data = [ (datetime( 2007, 10, 11), 'Peak of bull market'), (datetime( 2008, 3, 12), 'Bear Stearns Fails'), (datetime( 2008, 9, 15), 'Lehman Bankruptcy') ]

for date, label in crisis_data:
    ax.annotate(label, xy =(date, spx.asof(date) + 75), xytext =(date, spx.asof(date) + 225), 
                arrowprops = dict(facecolor ='black', headwidth = 10, width = 3,headlength = 4),
                horizontalalignment ='left', verticalalignment ='top')
    
# Zoom in on 2007-2010 
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600,1800])

ax.set_title('Important dates in the 2008-2009 financial crisis')

plt.show()                                                                                                                    
                                                                                                                     

# McKinney, Wes. Python for Data Analysis (p. 406). O'Reilly Media. Kindle Edition. 
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot( 1, 1, 1)

rect = plt.Rectangle(( 0.2, 0.75), 0.4, 0.15, color ='k', alpha = 0.3)
circ = plt.Circle(( 0.7, 0.2), 0.15, color ='b', alpha = 0.3)
pgon = plt.Polygon([[ 0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color ='g', alpha = 0.5)

ax.add_patch(rect) 
ax.add_patch(circ)
ax.add_patch(pgon)

plt.show()
```

# Scatter or Point Plots


```python
macro=pd.read_csv('data_sets/macrodata.txt')
macro.head()
```

```python
data=macro[['cpi','m1','tbilrate','unemp']]

transformed_data=np.log(data).diff().dropna()
transformed_data[-5:]
```

```python
sns.regplot('m1','unemp',data=transformed_data)
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

```python
sns.pairplot(transformed_data, diag_kind ='kde', plot_kws ={'alpha': 0.2})
```

```python
sns.pairplot(transformed_data, diag_kind ='kde',kind='reg')
```
