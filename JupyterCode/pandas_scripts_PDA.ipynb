```python
from numpy import nan as NA 
data = pd.Series([1, NA,3.5, NA,7]) 
print(data)
cleaned=data.dropna()
print(cleaned)
```

```python
data = pd.DataFrame([[ 1., 6.5, 3.], [1., NA, NA],[NA, NA, NA], [NA, 6.5, 3.]])
print(data,'\n')
cleaned=data.dropna()
print(cleaned)
```

```python
# drop if all are na in row (0) axis
cleaned=data.dropna(how='all')
print(cleaned)
```
