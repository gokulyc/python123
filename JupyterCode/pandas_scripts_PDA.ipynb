# Pivoting “Long” to “Wide” Format  

- A common way to store multiple time series in databases and CSV is in so-called long or stacked format.   
- Let’s load some example data and do a small amount of time series wrangling and other data cleaning.  


```python
data=pd.read_csv('data_sets/pydata-book-master_ch07_macrodata.csv')
data.head()
```

```python
periods=pd.PeriodIndex(year=data.year,quarter=data.quarter,name='date')
periods
```

```python
columns=pd.Index(['realgdp','infl','unemp'])
columns
```

```python
data=data.reindex(columns=columns)
```

```python
data
```

```python
data.index=periods.to_timestamp('D','end')
data
```

```python
data.stack()
```

```python
ldata=data.stack().reset_index().rename(columns={0:'value'})
ldata.head(10)
```

```python
ldata.shape
```

```python
pivoted=ldata.pivot('date','level_1','value')
pivoted
```

```python
ldata['value2']=np.random.randn(len(ldata))
ldata[:10]
```

```python
pivoted=ldata.pivot('date','level_1')
```

```python
pivoted
```

```python
pivoted['value'][:5]
```

```python
unstacked = ldata.set_index(['date', 'level_1']).unstack('level_1')
unstacked
```

# Pivoting “Wide” to “Long” Format


- An inverse operation to pivot for DataFrames is pandas.melt.

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'], 'A': [1, 2, 3],  'B': [4, 5, 6],  'C': [7, 8, 9]})
df
```

```python
melted=pd.melt(df,['key'])
melted
```

```python
reshaped=melted.pivot('key','variable','value')
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df,'key')
```

```python
pd.melt(df,id_vars=['key'],value_vars=['A','B'])
```

```python
pd.melt(df,value_vars=['A','B','C'])
```
